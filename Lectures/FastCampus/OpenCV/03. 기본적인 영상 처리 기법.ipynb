
  - src1: ì²« ë²ˆì§¸ ì˜ìƒ ë˜ëŠ” ìŠ¤ì¹¼ë¼
  - src2: ë‘ ë²ˆì§¸ ì˜ìƒ ë˜ëŠ” ìŠ¤ì¹¼ë¼
  - dst: ì°¨ì´ ì—°ì‚° ê²°ê³¼ ì˜ìƒ(ì°¨ì˜ìƒ)

````python
import sys
import numpy as np
import cv2
from matplotlib import pyplot as plt

src1 = cv2.imread('./data/images/lenna256.bmp', cv2.IMREAD_GRAYSCALE)
src2 = cv2.imread('./data/images/square.bmp', cv2.IMREAD_GRAYSCALE)

if src1 is None or src2 is None:
    print("Image load failed!")
    sys.exit()

dst1 = cv2.add(src1, src2, dtype=cv2.CV_8U)
dst2 = cv2.addWeighted(src1, 0.5, src2, 0.5, 0.0)
dst3 = cv2.subtract(src1, src2)
dst4 = cv2.absdiff(src1, src2)

plt.subplot(231), plt.axis('off'), plt.imshow(src1, 'gray'), plt.title('src1')
plt.subplot(232), plt.axis('off'), plt.imshow(src2, 'gray'), plt.title('src2')
plt.subplot(233), plt.axis('off'), plt.imshow(dst1, 'gray'), plt.title('add')
plt.subplot(234), plt.axis('off'), plt.imshow(dst2, 'gray'), plt.title('addWeighted')
plt.subplot(235), plt.axis('off'), plt.imshow(dst3, 'gray'), plt.title('substract')
plt.subplot(236), plt.axis('off'), plt.imshow(dst4, 'gray'), plt.title('absdiff')
plt.show()
````

## 06. íˆìŠ¤í† ê·¸ë¨ í‰í™œí™”

### íˆìŠ¤í† ê·¸ë¨ í‰í™œí™”(Histogram equalization) 
- íˆìŠ¤í† ê·¸ë¨ì´ ê·¸ë ˆì´ìŠ¤ì¼€ì¼ ì „ì²´ êµ¬ê°„ì—ì„œ ê· ì¼í•œ ë¶„í¬ë¡œ ë‚˜íƒ€ë‚˜ë„ë¡ ë³€ê²½í•˜ëŠ”

- íˆìŠ¤í† ê·¸ë¨ ê· ë“±í™”, ê· ì¼í™”, í‰íƒ„í™”

    ```py
    cv2.equalizeHist(src, dst=None) -> dst
    ```
    - src: ì…ë ¥ ì˜ìƒ. *ê·¸ë ˆì´ìŠ¤ì¼€ì¼ ì˜ìƒ.*
    - dst: ê²°ê³¼ ì˜ìƒ

### íˆìŠ¤í† ê·¸ë¨ í‰í™œí™”ë¥¼ ìœ„í•œ ë³€í™˜ í•¨ìˆ˜ êµ¬í•˜ê¸°
- íˆìŠ¤í† ê·¸ë¨ í•¨ìˆ˜ êµ¬í•˜ê¸°: $h(g) = N_g$
- ì •ê·œí™”ëœ íˆìŠ¤í† ê·¸ë¨ í•¨ìˆ˜ êµ¬í•˜ê¸°: $p(g) = \frac{h(g)}{w \times h}$
- *ëˆ„ì  ë¶„í¬ í•¨ìˆ˜(cdf)* êµ¬í•˜ê¸°: $cdf(g) = \sum_{0 \le i <g} P^{(i)}$
- ë³€í™˜ í•¨ìˆ˜: $dst(x, y) = round(cdf(src(x, y)) \times L_{max})$

## 08. íˆìŠ¤í† ê·¸ë¨ ì—­íˆ¬ì˜
### íˆìŠ¤í† ê·¸ë¨ ì—­íˆ¬ì˜(Histogram backprojection)
- ì˜ìƒì˜ ê° í”½ì…€ì´ ì£¼ì–´ì§„ íˆìŠ¤í† ê·¸ë¨ ëª¨ë¸ì— ì–¼ë§ˆë‚˜ ì¼ì¹˜í•˜ëŠ”ì§€ë¥¼ ê²€ì‚¬í•˜ëŠ” ë°©ë²•
- ì„ì˜ì˜ ìƒ‰ìƒ ì˜ì—­ì„ ê²€ì¶œí•  ë•Œ íš¨ê³¼ì 

### íˆìŠ¤í† ê·¸ë¨ ì—­íˆ¬ì˜ í•¨ìˆ˜
```py
cv2.calcBackProject(images, channels, hist, ranges, scale, dst=None) -> dst
```
- images: ì…ë ¥ ì˜ìƒ ë¦¬ìŠ¤íŠ¸
- channels: ì—­íˆ¬ì˜ ê³„ì‚°ì— ì‚¬ìš©í•  ì±„ë„ ë²ˆí˜¸ ë¦¬ìŠ¤íŠ¸
- hist: ì…ë ¥ íˆìŠ¤í† ê·¸ë¨ (numpy.ndarray)
- ranges: íˆìŠ¤í† ê·¸ë¨ ê° ì°¨ì›ì˜ ìµœì†Ÿê°’ê³¼ ìµœëŒ“ê°’ìœ¼ë¡œ êµ¬ì„±ëœ ë¦¬ìŠ¤íŠ¸
- scale: ì¶œë ¥ ì—­íˆ¬ì˜ í–‰ë ¬ì— ì¶”ê°€ì ìœ¼ë¡œ ê³±í•  ê°’
- dst: ì¶œë ¥ ì—­íˆ¬ì˜ ì˜ìƒ. ì…ë ¥ ì˜ìƒê³¼ ë™ì¼ í¬ê¸°, cv2.CV_8U.

````python
import sys
import numpy as np
import cv2

# ----------------------------------------------------------------------------------
# ì…ë ¥ ì˜ìƒì—ì„œ ROIë¥¼ ì§€ì •í•˜ê³ , íˆìŠ¤í† ê·¸ë¨ ê³„ì‚°
src = cv2.imread('./data/images/cropland.png')

if src is None:
    print('Image load failed!')
    sys.exit()

x, y, w, h = cv2.selectROI(src)

####################################################################################
# íˆìŠ¤í† ê·¸ë¨ì„ ê³„ì‚°í•˜ëŠ” ì½”ë“œ

src_ycrcb = cv2.cvtColor(src, cv2.COLOR_BGR2YCrCb)         # YCrCb ColorSpaceë¡œ ë³€í™˜
crop = src_ycrcb[y:y+h, x:x+w]                # ì‚¬ìš©ìê°€ ì„ íƒí•œ ì‚¬ê°í˜• ì˜ì—­ì˜ ë¶€ë¶„ ì˜ìƒ

channels = [1, 2]                              # 0ë²ˆ ì±„ë„ì¸ Y(ë°ê¸°)ì„±ë¶„ì€ ì‚¬ìš©í•˜ì§€ ì•ŠìŒ 
cr_bins = 128                 # cr, cbëŠ” ì›ë˜ 256ì„ ì‚¬ìš©í•˜ëŠ” ê²ƒì´ ì •ìƒì´ë‚˜, 128ë¡œ ë‹¨ìˆœí™”
cb_bins = 128
histSize = [cr_bins, cb_bins]
cr_range = [0, 256]                                              # 1ë²ˆ ì±„ë„ crì˜ ë²”ìœ„
cb_range = [0, 256]                                              # 2ë²ˆ ì±„ë„ cbì˜ ë²”ìœ„
ranges = cr_range + cb_range

hist = cv2.calcHist([crop], channels, None, histSize, ranges)       # íˆìŠ¤í† ê·¸ë¨ ê³„ì‚°
####################################################################################
hist_norm = cv2.normalize(cv2.log(hist+1), None, 0, 255, cv2.NORM_MINMAX, cv2.CV_8U)
# ----------------------------------------------------------------------------------

# ----------------------------------------------------------------------------------
# ì…ë ¥ ì˜ìƒ ì „ì²´ì— ëŒ€í•´ íˆìŠ¤í† ê·¸ë¨ ì—­íˆ¬ì˜

# ê³„ì‚°í•œ íˆìŠ¤í† ê·¸ë¨ì„ ì´ìš©í•´ calcBackProject()ë¡œ ì›í•˜ëŠ” ì˜ì—­ì˜ ì‚¬ì§„ë§Œ ê³¨ë¼ëƒ„
backproj = cv2.calcBackProject([src_ycrcb], channels, hist, ranges, 1)
dst = cv2.copyTo(src, backproj)

cv2.imshow('backproj', backproj)
cv2.imshow('hist_norm', hist_norm)
cv2.imshow('dst', dst)

cv2.waitKey()
cv2.destroyAllWindows()
# ----------------------------------------------------------------------------------
````

````python
import sys
import numpy as np
import cv2

# ë…¹ìƒ‰ ë°°ê²½ ë™ì˜ìƒ
cap1 = cv2.VideoCapture('./data/videos/woman.mp4') 

if not cap1.isOpened():
    print('video open failed!')
    sys.exit()

# ë¹„ì˜¤ëŠ” ë°°ê²½ ë™ì˜ìƒ
cap2 = cv2.VideoCapture('./data/videos/raining.mp4')

if not cap2.isOpened():
    print('video open failed!')
    sys.exit()

# ë‘ ë™ì˜ìƒì˜ í¬ê¸°, FPSëŠ” ê°™ë‹¤ê³  ê°€ì •
w = round(cap1.get(cv2.CAP_PROP_FRAME_WIDTH))
h = round(cap1.get(cv2.CAP_PROP_FRAME_HEIGHT))
frame_cnt1 = round(cap1.get(cv2.CAP_PROP_FRAME_COUNT))
frame_cnt2 = round(cap2.get(cv2.CAP_PROP_FRAME_COUNT))
print(f'w Ã— h: {w} Ã— {h}')
print('frame_cnt1:', frame_cnt1)
print('frame_cnt2:', frame_cnt2)

fps = cap1.get(cv2.CAP_PROP_FPS)
delay = int(1000 / fps)

# ì¶œë ¥ ë™ì˜ìƒ ê°ì²´ ìƒì„±
fourcc = cv2.VideoWriter_fourcc(*'DIVX')
out = cv2.VideoWriter('./data/output.avi', fourcc, fps, (w, h))

# í•©ì„± ì—¬ë¶€ í”Œë˜ê·¸
do_composit = False

# ì „ì²´ ë™ì˜ìƒ ì¬ìƒ
while True:
    ret1, frame1 = cap1.read()

    if not ret1:
        break
    
    # do_composit í”Œë˜ê·¸ê°€ Trueì¼ ë•Œì—ë§Œ í•©ì„±
    if do_composit:
        ret2, frame2 = cap2.read()

        if not ret2:
            break
        
        # frame2ì˜ í¬ê¸°ë¥¼ ì…ë ¥ ë°›ì€ ì˜ìƒê³¼ ë™ì¼í•˜ê²Œ ë³€ê²½
        frame2 = cv2.resize(frame2, (w, h))

        # HSV ìƒ‰ ê³µê°„ì—ì„œ ë…¹ìƒ‰ ì˜ì—­ì„ ê²€ì¶œí•˜ì—¬ í•©ì„±
        hsv = cv2.cvtColor(frame1, cv2.COLOR_BGR2HSV)
        mask = cv2.inRange(hsv, (50, 150, 0), (70, 255, 255))
        cv2.copyTo(frame2, mask, frame1)

    cv2.imshow('frame', frame1)
    key = cv2.waitKey(delay)

    # ìŠ¤í˜ì´ìŠ¤ë°”ë¥¼ ëˆ„ë¥´ë©´ do_composit í”Œë˜ê·¸ë¥¼ ë³€ê²½
    if key == ord(' '):
        do_composit = not do_composit
    elif key == 27:
        break

cap1.release()
cap2.release()
out.release()
cv2.destroyAllWindows()
````
